```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os

# Load the model
model = load_model('/content/best_model_2.h5', compile=False)

# Image file names (you can update this list)
image_names = ['IMG_10.jpg', 'IMG_11.jpg', 'IMG_12.jpg','IMG_1.jpg','IMG_100.jpg','IMG_113.jpg','IMG_300.jpg','IMG_25.jpg','IMG_250.jpg','IMG_99.jpg']

# Base paths
image_base_path = '/content/crowd dataset/test_data/images/'
gt_base_path = '/content/test density/'

# Loop over the 3 images
for image_name in image_names:
    # Load image
    img_path = os.path.join(image_base_path, image_name)
    img = load_img(img_path, target_size=(768, 1024))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    # Predict density map
    prediction = model.predict(img_array)
    predicted_count = np.sum(prediction)

    # Load ground truth density map
    gt_name = 'GT_' + image_name.replace('.jpg', '.npy')
    gt_path = os.path.join(gt_base_path, gt_name)
    true_density = np.load(gt_path)
    true_count = np.sum(true_density)

    # Display results
    print(f'\n Image: {image_name}')
    print(f' Predicted Count: {predicted_count:.2f}')
    print(f'Ground Truth Count: {true_count:.2f}')

    # Plot
    plt.figure(figsize=(9, 3))

    # Original image
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title("Original Image")
    plt.axis('off')

    # Ground Truth
    plt.subplot(1, 3, 2)
    plt.imshow(true_density, cmap='jet')
    plt.title(f"Ground Truth Density Map\nCount: {true_count:.2f}")
    plt.axis('off')

    # Prediction
    plt.subplot(1, 3, 3)
    plt.imshow(prediction[0, :, :, 0], cmap='jet')
    plt.title(f"Predicted Density Map\nCount: {predicted_count:.2f}")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
